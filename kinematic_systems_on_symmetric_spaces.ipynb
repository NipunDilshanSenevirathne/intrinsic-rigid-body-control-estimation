# Overview: Geometric Virtual Structure and Tracking Control for Multi Agent Systems on Lie Groups

This repository summarizes the geometric control framework for a network of fullyactuated simple mechanical systems evolving on a Lie group $ G $.  
The framework extends the **AGLES-PID** and **virtual-structure synchronization** methodology to multi-agent systems.

---

## 1. System Setup

### Agents and Virtual Systems
We consider $ N $ agents
\begin{align}
\mathscr{A} = \{A_1, A_2, \dots, A_N\},
\end{align}
each modeled as a fully actuated simple mechanical system on a Lie group $ G $ with Lie algebra $ \mathfrak{g} $ and dual $ \mathfrak{g}^* $.

Each agent $A_i$ and its virtual counterpart $V_i$ evolve as:
\begin{align}
(g_i, \pi_i) \in G \times \mathfrak{g}^*, \qquad
(g_{v_i}, \pi_{v_i}) \in G \times \mathfrak{g}^*,
\end{align}
with left–invariant kinetic energy metrics defined by constant inertia tensors:
\begin{align}
\mathbb{I}_i : \mathfrak{g} \to \mathfrak{g}^*, \qquad
\mathbb{I}_{v_i} : \mathfrak{g} \to \mathfrak{g}^*.
\end{align}

---

## 2. Right–Invariant Dynamics

In momentum coordinates, each system evolves as:
\begin{align}
\dot{g} = \omega \cdot g, \qquad
\dot{\pi} = f^e + f^u,
\end{align}
where
- $ \omega = \mathbb{I}^{-1}\pi \in \mathfrak{g} $ is the body velocity,
- $ f^e \in \mathfrak{g}^* $ is an external (disturbance) force, and
- $ f^u \in \mathfrak{g}^* $ is the fully actuated control input.

For virtual systems $V_i$, $f^e_{v_i} \equiv 0$.

---

## 3. Agent Control Strategy

Each agent executes a **three–step strategy**:

1. **Local tracking objective**  
   Define
   \begin{align}
   g_r(t) = \bar{g}_i g_{v_i}(t),
   \end{align}
   where $ \bar{g}_i \in G $ is constant.  
   In $ SE(3) $, this represents maintaining a fixed pose offset relative to $V_i$.

2. **Agent tracking problem**  
   Choose $ f_i^u $ so that  
   \begin{align}
   \lim_{t \to \infty} g_i(t) = g_r(t).
   \end{align}
   If $ g_r(t) = \bar{g}_i g_{v_i}(t) $, then $ A_i $ converges to configuration $ \bar{g}_i $ relative to $V_i$.

3. **Virtual system synchronization**  
   Choose a communication strategy and $ f^u_{v_i} $ so that  
   \begin{align}
   \lim_{t \to \infty} g_{v_i}(t) = g_v(t), \quad \forall i,
   \end{align}
   ensuring asymptotic consensus of all virtual systems.

---

## 4. Error Dynamics

Define the **right–invariant tracking error**
\begin{align}
e = g_v g_i^{-1}.
\end{align}
Then:
\begin{align}
\dot{e} &= \omega_e \cdot e, \qquad
\omega_e &= \omega_v - \operatorname{Ad}_e \omega_i.
\end{align}

Define the **error momentum**
\begin{align}
\pi_e = \operatorname{Ad}_{g_i}^* \mathbb{I}\operatorname{Ad}_{g_v^{-1}}\omega_e
       = \operatorname{Ad}_{e^{-1}}^*\pi_v - \pi_i.
\end{align}

The **momentum error dynamics** (with right–invariant kinematics) are:
\begin{align}
\boxed{
\dot{\pi}_e
= \operatorname{Ad}_{e^{-1}}^*\dot{\pi}_v
- \dot{\pi}_i
+ \operatorname{ad}_{\omega_e}^*\operatorname{Ad}_{e^{-1}}^*\pi_v.
}
\end{align}

---

## 5. AGLES–PID Control Law

Let $ f_e : G \to \mathbb{R} $ be a polar Morse function with unique minimum at the identity.  
Define
\begin{align}
df_e = \pi_e \cdot e, \qquad
\dot{\pi}_I = \pi_e.
\end{align}

The **control law** is:
\begin{align}
f_i^u =
\left(
\operatorname{Ad}_{e^{-1}}^*\dot{\pi}_r
+ \operatorname{ad}_{\omega_e}^*\operatorname{Ad}_{e^{-1}}^*\pi_r
- f_i^e
\right)
- k_p\pi_e - k_d\pi_e - k_I\pi_I.
\end{align}

The **closed-loop error dynamics** are:
\begin{align}
\dot{e} = \omega_e \cdot e,\\
\dot{\pi}_I = \pi_e,\\
\dot{\pi}_e = -k_p\pi_e - k_d\pi_e - k_I\pi_I.
\end{align}
> “The error dynamics do not get any simpler or more straightforward than this.”

---

## 6. Synchronization Strategies

### Centralized
A global controller broadcasts $ g_v(t) $;  each agent uses its AGLES–PID law to track $ g_v(t) $.  The broadcast can be encoded or key–protected.

### Decentralized
Each agent communicates with neighbors to share $(g_{v_i}(t), \pi_{v_i}(t))$ and applies a consensus-based control $ f^u_{v_i} $ to achieve
\begin{align}
\lim_{t\to\infty} g_{v_i}(t) = g_v(t), \quad \forall i.
\end{align}

## 7. Mathematical Conventions

**Coadjoint action:**  
$$
\begin{align}
\langle \operatorname{Ad}_g^*\mu, \zeta \rangle = \langle \mu, \operatorname{Ad}_{g^{-1}}\zeta \rangle
\end{align}
$$

**Right–invariant kinematics:**  
$$
\begin{align}
\dot{g} = \omega g
\end{align}
$$

**Derivative rule:**  
$$
\begin{align}
\frac{d}{dt}(\operatorname{Ad}_{e^{-1}}^*\pi)
= \operatorname{Ad}_{e^{-1}}^*\dot{\pi}
+ \operatorname{ad}_{\omega_e}^*\operatorname{Ad}_{e^{-1}}^*\pi
\end{align}
$$

# Geometric Virtual Structure and Tracking Control for Multi Agent Systems on Lie Groups

## Lie group preliminaries

Let $G$ be a Lie group with Lie algebra $\mathfrak{g}$ and dual $\mathfrak{g}^*$. Define the inertia tensor $\mathbb{I}:\mathfrak{g}\mapsto \mathfrak{g}^*$ by the relationship
\begin{align}
\left\langle\mathbb{I}\zeta,\eta\right\rangle
\triangleq \frac{1}{2}\left\langle\left\langle\zeta,\eta\right\rangle \right\rangle
\end{align}
for any $\zeta,\eta\in \mathfrak{g}$.

Recall that $\operatorname{Ad}_{g}^*:\mathfrak{g}^*\mapsto \mathfrak{g}^*$
is defined by
\begin{align}
\left\langle\operatorname{Ad}_{g}^*\mu\,,\,\zeta\right\rangle=
\left\langle\mu\,,\,\operatorname{Ad}_{g^{-1}}\zeta\right\rangle
\end{align}
for any $\zeta\in \mathfrak{g}$ and $\mu\in \mathfrak{g}^*$.
Also define the $g$ depedned inertia tensor $\mathbb{I}_g:\mathfrak{g}\mapsto \mathfrak{g}^*$ by the relationship
\begin{align}
\left<\mathbb{I}_g\zeta,\eta\right>
\triangleq \frac{1}{2}\left\langle\left\langle \operatorname{Ad}_{g}\zeta\,,\,\operatorname{Ad}_{g}\eta\right\rangle \right\rangle
=\left\langle \mathbb{I}\operatorname{Ad}_{g}\zeta\:,\:\operatorname{Ad}_{g}\eta\right\rangle
=\left\langle \operatorname{Ad}_{g^{-1}}^*\mathbb{I}\operatorname{Ad}_{g}\zeta\:,\:\eta\right\rangle.  
\end{align}
From which we have
\begin{align}
\mathbb{I}_g&\triangleq \operatorname{Ad}_{g^{-1}}^*\mathbb{I}\operatorname{Ad}_{g}
\end{align}

Let $g\in G$,  and $\pi\in \mathfrak{g}^*$. Then define $\Pi\in \mathfrak{g}^*$ as
\begin{align}
\Pi&=\operatorname{Ad}_{g^{-1}}^{*}\, \pi.
\end{align}
Using $\mathbb{I}$ define $\Omega\in \mathfrak{g}$ by
\begin{align}
\Omega&=\mathbb{I}^{-1}\Pi.
\end{align}

Let $\omega\in \mathfrak{g}$ be such that
\begin{align}
\omega&=\operatorname{Ad}_{g}\, \Omega.
\end{align}

Now we have
\begin{align}
\left<\pi,\omega\right>&=\left<\operatorname{Ad}_{g}^{*}\Pi\,,\,\operatorname{Ad}_{g}\Omega\right>=\left<\Pi,\Omega\right>= \frac{1}{2}\left\langle\left\langle\Omega,\Omega\right\rangle \right\rangle,
\end{align}
and
\begin{align}
\omega&=\operatorname{Ad}_{g}\, \Omega=\operatorname{Ad}_{g}\, \mathbb{I}^{-1}\Pi=\left(\operatorname{Ad}_{g}\mathbb{I}^{-1}\operatorname{Ad}_{g^{-1}}^*\right)\pi,
\end{align}
Hence we also have
\begin{align}
\pi&=\left(\operatorname{Ad}_{g}^*\mathbb{I}\operatorname{Ad}_{g^{-1}}\right)\omega,
\end{align}

---

When we refer to rigid body attitude motion, $G=SO(3)$, and then $\Omega$ is body angular velocity, $\Pi$ is body angular momentum, $\omega=R\Omega$ is spatial angular velocity, $\pi=R\Pi$ is spatial angular momentum, $\mathbb{I}_R=R^T\mathbb{I}R$ is the locked inertia tensor and we have
\begin{align}
\left<\pi,\omega\right>&=\left<R\Pi\,,\,R\Omega\right>=\left<\Pi,\Omega\right>= \frac{1}{2}\left\langle\left\langle\Omega,\Omega\right\rangle \right\rangle
\end{align}


## Problem Definition

Consider the problem of $N$ agents $\mathscr{A}=\{A_1,A_2,\cdots,A_N\}$. Each agent $A_i$ is assumed to be a fully actuated simple mechanical system on a Lie group $G$ with Lie algebra $\mathfrak{g}$ and dual $\mathfrak{g}^*$ and left invariant kinetic energy metric defined by a constant inertia tensor $\mathbb{I}_i$. The state of $A_i$ is defined by $(g_i,\pi_i)\in G\times \mathfrak{g}^*$.

Consider the virtual fully actuated simple mechanical system (virtual frame, virtual structure, etc), $V_i$, that is also defined on the same Lie group $G$ with Lie algebra $\mathfrak{g}$. Let its inertia tensor be defined by $\mathbb{I}_{v_i}$. The state of the virtual system $V_i$ is defined by $(g_{v_i},\pi_{v_i})\in G\times \mathfrak{g}^*$.

Using the right invariant (momentum) representation
we have that the dynamics of the pair $(A_{v_i},A_i)$ evolve according to

\begin{align}
\dot{g}_{v_i}&=\omega_{v_i} \cdot g_{v_i},\\
\dot{\pi}_{v_i}&=f^e_{v_i}+f^u_{v_i}
\end{align}
where $f^e_{v_i}\in \mathfrak{g}^*$ is the external force on the system and
$f^u_{v_i}\in \mathfrak{g}^*$ is the fully actuated control force. We may set $f^e_{v_i}\equiv 0$.

The agent dynamics are assumed to be
\begin{align}
\dot{g}_i&=\omega_i \cdot g_i,\\
\dot{\pi}_i&=f^e_i+f^u_i
\end{align}
where $f^e_i\in \mathfrak{g}^*$ is the external force on the system and
$f^u_i\in \mathfrak{g}^*$ is the fully actuated control force.

## Agent Action


**Each agent executes a three-step strategy**


* **Define a local tracking objective:** Costruct $g_r(t)$ so that it is a function of $g_{v_i}(t)$. For isntance, the choice $g_r(t)=\bar{g}_ig_{v_i}(t)$ characterizes the positioning of $A_i$ at a configuration $\bar{g}_i$ with respect to the virtual $V_i$.

* **Agent local tracking problem:** Choose $f^u_i$ so that $\lim_{t\to \infty} g_i(t)=g_r(t)$. This again if we choose $g_r(t)=\bar{g}_ig_{v_i}(t)$ ensures that $\lim_{t\to \infty}g_i(t)=\bar{g}_i$ with respect to $V_i$.

* **Virtual system synchronization through communication:** Choose a communication strtegy and $f^u_{v_i}$ so that $\lim_{t\to \infty} g_{v_i}(t)={g}_{v}(t)$ for all $i$. This again if each agent chooses $g_r(t)=\bar{g}_ig_{v_i}(t)$ then this automatically ensures that all agents occupy constant configurations with respect to a globally consistant reference frame ${g}_{v}(t)$.


### Defining the local tracking objective

Let
\begin{align}
g_r(t)=\bar{g}_ig_{v_i}(t),
\end{align}
where $\bar{g}_i\in G$ is a constant. With respect to $SE(3)$ this tracking reference represents the objective of $A_i$ positioning it self at $\bar{g}_i$ with respect to $V_i$.

The cosntant $\bar{g}_i$ is only known to $A_i$ and may even be commanded by a centralized controller.

### Agent local tracking problem

#### Local tracking error dynamics

Given a sufficiently smooth trajectory $g_r(t)$, possibly chosen as above, define $\omega_r(t)=\dot{g}_rg_r^{-1}$ and $\pi_r(t)=\left(\operatorname{Ad}_{g_r}^*\mathbb{I}\operatorname{Ad}_{g_r^{-1}}\right)\omega_r$.

Define the right-invariant tracking error $e=g_rg_i^{-1}$.

Differentiating we have
\begin{align}
\dot{e}&=\omega_e\cdot e
\end{align}
Where
\begin{align}
\omega_e&\triangleq\left(\omega_r-\operatorname{Ad}_{e}\omega_i\right)
\end{align}
Define
\begin{align}
\pi_e&\triangleq
\left(\operatorname{Ad}_{g_i}^*\mathbb{I}\operatorname{Ad}_{g_r^{-1}}\right)\omega_e=
\left(\operatorname{Ad}_{g_i}^*\mathbb{I}\operatorname{Ad}_{g_r^{-1}}\right)\left(\omega_{v_i}-\operatorname{Ad}_{e}\omega_i\right)=
\operatorname{Ad}_{e^{-1}}^*\pi_r-\pi_i.
\end{align}

Differentiating we have
\begin{align}
\dot{\pi}_e&=
\operatorname{Ad}_{e^{-1}}^*\dot{\pi}_r-\dot{\pi}_i+
\operatorname{ad}_{\omega_e}^*\operatorname{Ad}_{e^{-1}}^*\pi_r\\
&=\operatorname{Ad}_{e^{-1}}^*\dot{\pi}_r+
\operatorname{ad}_{\omega_e}^*\operatorname{Ad}_{e^{-1}}^*\pi_r-
(f^e_i+f^u_i)
\end{align}

#### Intinsic almost global locally exponentially stable PID - local virtual system synchronization:


Let $f_e:G\mapsto \mathbb{R}$ be a polar Morse function on $G$ with a unique minimum at the identity. Define $\pi_e$ by
\begin{align}
df_e&=\pi_e\cdot e
\end{align}
where $df_e$ is the straight forward differential of $f_e(e)$ at $e$.

\begin{align}
\dot{\pi_I} &= \pi_e\\
f^u_i&=\left(\operatorname{Ad}_{e^{-1}}^*\dot{\pi}_r+
\operatorname{ad}_{\omega_e}^*\operatorname{Ad}_{e^{-1}}^*\pi_r-f^e_i\right)
-k_p\pi_e-k_d\pi_e-k_I\pi_I
\end{align}

#### Controlled error dynamics

\begin{align}
\dot{e}&=\omega_e\cdot e,\\
\dot{\pi}_I&=\pi_e,\\
\dot{\pi}_e&=-k_p\pi_e-k_d\pi_e-k_I\pi_I
\end{align}

**The error dynamics do not get any simpler or more straight forward than this.**

### Virtual system synchronization through communication

The objective here is to choose an appropriate centralized/decentralized communication strategy so that all agent virtual systems $V_i$ are assymptotically synchronized.

* **Centralized Approach:** A centralized controller globally boradcasts $g_v(t)$ and each agent uses an AGLES PID controller to ensure that $\lim_{t\to \infty}g_{v_i}(t)=g_v(t)$. This  $g_v(t)$ can be encoded with a key only known to agents.

* **Decentalized Approcah:** Each agent communicates with it neighbors their own $(g_{v_i}(t),\pi_{v_i}(t))$ and a syncronization controller for $f^y_{v_i}$ is chosen. Again $g_{v_i}$ can be appropriately coded.

# Flocking

---

## 🧭 1. The general form of flocking / alignment equations

In most continuous-time formulations, the system looks like:

\begin{align}
\dot{p}_i & = v_i, \\
\dot{v}_i & = \sum_{j} a_{ij}(p_i, p_j) (v_j - v_i) - \nabla_{p_i}V(p_1,\ldots,p_N),
\end{align}


where:

* $V$ is the **potential function** (cohesion/repulsion term),
* $a_{ij}$ are **interaction weights** (e.g. decaying with distance),
* velocities may have **fixed magnitude** $(|v_i| = v_0)$ or be free.

These dynamics may be viewed as the gradient flow of a **networked mechanical system** with potential and dissipative coupling.

When restricted to headings $h_i = v_i/|v_i|$, you get the **orientation dynamics**:

\begin{align}
\dot{h}_i &= \sum_j a_{ij}(p_i, p_j) P_{h_i}(h_j - h_i),\\
\quad P_{h_i} &= I - h_i h_i^\top.
\end{align}

That is a projected gradient system on $S^2$.

## ⚙️ 2. Theoretical results by model family

### **(a) Cucker–Smale (2007–present) — global convergence conditions**

This is the most complete theoretical treatment of alignment-type systems.

For
\begin{align}
\dot{p}_i &= v_i, \\
\dot{v}_i &= \frac{1}{N} \sum_{j=1}^N \psi(|p_j - p_i|)(v_j - v_i),
\end{align}
they proved:

* **Flocking theorem:**
  If the communication kernel, $\psi(r)$, decays slower than $1/r^2$,
  then **asymptotic velocity consensus** and **bounded group diameter** occur globally.

  Specifically, if $\int_0^\infty \psi(s),ds = \infty$,
  then velocities converge: $(v_i(t) - v_j(t) \to 0)$, and positions remain bounded.

* **Quantitative rate:**
  Velocity differences decay exponentially when $\psi$ is bounded away from 0.
  If $\psi(r) = (1+r^2)^{-\beta}$, convergence occurs for $\beta \le 1/2$.

* **Mean-field (kinetic) limit:**
  PDE limit is a **Vlasov-type equation** for density $f(x,v,t)$.
  Flocking corresponds to monokinetic convergence $f(x,v,t) \to \rho(x,t)\delta(v - v_\infty)$.

🧩 *Mathematical tools used*: Lyapunov analysis, Grönwall inequalities, coercivity of alignment operator, kinetic theory.
Main references:

* Cucker & Smale (IEEE TAC, 2007; CPAM, 2007)
* Ha & Tadmor (2008, 2009)
* Carrillo, Fornasier, Toscani (2010, 2013)

---

### **(b) Olfati–Saber (2006) — Lyapunov analysis of flocking potential**

For the continuous-time flocking model with potential (V):

\begin{align}
\dot{p}_i & = v_i, \\
\dot{v}_i & = -\nabla_{p_i} V(p) - c\sum_j a_{ij}(v_i - v_j),
\end{align}

Olfati-Saber proved:

* **Boundedness of inter-agent distances** under appropriate potential parameters,
* **Exponential convergence of velocities** to a common value,
* **Local asymptotic stability** of cohesive motion equilibria.

He defined a **flocking potential**
\begin{align}
V(p) = \frac{1}{2}\sum_{i,j} \phi(|p_i - p_j|),
\end{align}
where $\phi(r)$ has a unique minimum at desired spacing.

Using a composite Lyapunov function
\begin{align}
\mathscr{V} = V(p) + \frac{1}{2}\sum_i |v_i - v_c|^2,
\quad v_c = \frac{1}{N}\sum_i v_i,
\end{align}
he derived explicit exponential convergence rates for $\dot{\mathscr{V}} \le -c |v_i - v_j|^2$.

### **(c) Flocking and consensus on manifolds**

For **heading consensus on (S^2)** or attitude synchronization on (SO(3)):

\begin{align}
\dot{R}_i &= R_i \widehat{\Omega}_i,\\
\Omega_i & = -k \sum_{j} a_{ij} \operatorname{skew}(R_i^\top R_j).
\end{align}

Results (Sarlette, Sepulchre, Markdahl, Tron, 2009-2020):

* **Almost-global convergence** to synchronized attitudes (except for a measure-zero set of initial conditions, due to topological constraints on $SO(3$)).
* **Convergence rates** derived using Morse potentials $f(R_i^\top R_j) = \operatorname{tr}(I - R_i^\top R_j)$.
* **Linearization near consensus** yields exponential convergence with rate $k\lambda_2(L)$, where $L$ is the Laplacian of the communication graph.

For headings $h_i \in S^2$:

\begin{align}
\dot{h}_i = -k \sum*{j} a_{ij} P_{h_i}(h_i - h_j),
\end{align}
where $P_{h_i}$ projects to tangent space.
Results (Markdahl, Gonçalves, Thunberg 2018, 2020):

* **Almost-global convergence on $S^2$** with precise characterization of stable “splay” states.
* **Explicit potential landscape** (critical points correspond to synchronized or antipodal configurations).
* **Quantitative rates** from linearized Laplacian flow on tangent spaces.

---

### **(d) Stochastic and mean-field analysis**

When noise or diffusion terms are added:

\begin{align}
d v_i &= \sum_j a_{ij}(v_j - v_i),dt + \sigma,dW_i(t),
\end{align}
or for manifold-valued states:
\begin{align}
d R_i = R_i (\Omega_i,dt + \sigma \circ dW_i).
\end{align}

Then:

* **Mean-square convergence** and **stationary distribution** analysis are performed using Itô or Stratonovich calculus on Lie groups.
* Convergence rate depends on $\sigma$ (noise intensity) and graph algebraic connectivity $\lambda_2(L)$.

References:

* Wu & Tron (CDC 2020): “Stochastic consensus on manifolds.”
* Li & Egerstedt (IEEE TAC 2014): “Stochastic consensus on compact manifolds.”

## 🧠 3. Intuitive interpretation for flocking

The rigorous analysis shows that:

* The **potential ensures cohesion**, bounding relative positions.
* The **alignment term** is a *contracting operator* on velocities or headings.
* If information flow (graph connectivity) is maintained, the system acts as a **gradient-like flow** with an invariant set corresponding to consensus.

These results mathematically prove what early heuristic models only described qualitatively — i.e., flocking is the **stable attractor** of the coupled gradient–damping dynamics.

## 1 Rigid-Body Model on SE(3)

For each agent $A_i$:

\begin{align}
g_i &=
\begin{bmatrix}
R_i & x_i \\
0 & 1
\end{bmatrix}
\in SE(3), \qquad
R_i\!\in SO(3),\;
x_i\!\in\mathbb{R}^3.
\end{align}


- **Lie algebra:**  
$\mathfrak{se}(3)=\{(\Omega_i,v_i)\}$ with  
$\widehat{\Omega}_i\in\mathfrak{so}(3)$, $v_i\in\mathbb{R}^3$.

\begin{align}
\operatorname{Ad}_{g_i}
\begin{bmatrix}\Omega\\v\end{bmatrix}
&=
\begin{bmatrix}
R_i\Omega \\ R_i v + (R_i\Omega)\times x_i
\end{bmatrix},\qquad
\operatorname{Ad}_{g_i}^*
\begin{bmatrix}\Pi\\p\end{bmatrix}
=
\begin{bmatrix}
R_i\Pi + x_i\times R_i p\\
R_i p
\end{bmatrix}.
\end{align}

\begin{align}
\mathbb{I}_i&=
\begin{bmatrix}

\end{align}

## 4 · Right-Invariant Error

\begin{align}
e_i &= g_{v_i} g_i^{-1}
=
\begin{bmatrix}
R_e & x_e\\
0 & 1
\end{bmatrix},
\qquad
R_e = R_{v_i} R_i^\top,\;
x_e = x_{v_i} - R_e x_i.
\end{align}

**Error kinematics**

$$
\begin{aligned}
\dot{e}_i &= \omega_{e_i}\cdot e_i,\\
\omega_{e_i} &= \omega_{v_i}-\operatorname{Ad}_{e_i}\omega_i.
\end{aligned}
$$

**Error momentum**

$$
\pi_{e_i}
= \operatorname{Ad}_{g_i}^*\mathbb{I}_i\operatorname{Ad}_{g_{v_i}^{-1}}\omega_{e_i}
= \operatorname{Ad}_{e_i^{-1}}^*\pi_{v_i}-\pi_i.

**Momentum derivative identity**

$$
\frac{d}{dt}\!\left(\operatorname{Ad}_{e_i^{-1}}^*\pi_{v_i}\right)
= \operatorname{Ad}_{e_i^{-1}}^*\dot{\pi}_{v_i}
+ \operatorname{ad}_{\omega_{e_i}}^*\operatorname{Ad}_{e_i^{-1}}^*\pi_{v_i}.
$$

Hence

$$
\dot{\pi}_{e_i}
= \operatorname{Ad}_{e_i^{-1}}^*\dot{\pi}_{v_i}
- \dot{\pi}_i
+ \operatorname{ad}_{\omega_{e_i}}^*\operatorname{Ad}_{e_i^{-1}}^*\pi_{v_i}.
$$

+\tfrac12 k_x\|x\|^2$).

=

Closed-loop **error dynamics**

# Group Actions


Given a smooth and proper action of a Lie group $G$ on a smooth manifold $M$ is denoted by
\begin{align}
\Phi: G \times M \to M, \qquad
(g,z) \mapsto \Phi_g(z).
\end{align}
Define the **orbit** of each point $ z \in M $ as
\begin{align}
\mathscr{O}(z) = \{\, \Phi_g(z) \mid g \in G \,\}=[z].
\end{align}

The set of all orbits forms the **quotient space**
\begin{align}
M / G = \{\, \mathscr{O}(z) \mid z \in M \,\}.
\end{align}

## Free actions



If the action is **free** the quotient space $M / G $ inherits a smooth-manifold structure.

Under these conditions, the projection
\begin{align}
\pi: M \to M / G,\qquad
z \mapsto \mathscr{O}(z)=[z],
\end{align}
defines a **smooth submersion**, making
\begin{align}
M \xrightarrow{\ \pi\ } M / G
\end{align}
a **principal $G$-bundle**.

At each point $ z \in M $, the tangent space decomposes as
\begin{align}
T_z M = \mathscr{V}_z \oplus \mathscr{H}_z,
\end{align}
where

- the **vertical subspace** $ \mathscr{V}_z $ is tangent to the group orbit:
\begin{align}
\mathscr{V}_z = \{\, \xi_M(z) \mid \xi \in \mathfrak{g} \,\},
\end{align}
with $ \xi_M(z) = \frac{d}{dt}\big|_{t=0} \Phi_{\exp(t\xi)}(z) $ the infinitesimal generator.

- the **horizontal subspace** $ \mathscr{H}_z $ is a complementary subspace satisfying
\begin{align}
T_z M = \mathscr{V}_z \oplus \mathscr{H}_z.
\end{align}

---

A consitant unique choice of $\mathscr{H}_z$ for all $z\in M$ is possible if one has a $G$-invariant metric on $M$ or there exists

## Transitive actions

When the action is **transitive**, there is only one orbit: the entire manifold $M$. Consequently, the quotient space $M/G$ collapses to a single point, making the projection $\pi: M \to M/G$ mathematically trivial.

However, a new and powerful structure emerges *inside* $M$ itself. Instead of $M$ being the total space of a bundle over a quotient, $M$ becomes the quotient of the group $G$.

Here is the breakdown of what you get:

### The Isomorphism Theorem
Because the action is transitive, every point in $M$ can be reached from any other point. If you pick an arbitrary reference point $z_0 \in M$, the manifold is diffeomorphic to the quotient of the group by the stabilizer of that point:
$$
M \cong G / H
$$
where $H = \text{Stab}(z_0) = \{ g \in G \mid g \cdot z_0 = z_0 \}$ is the **isotropy subgroup** (stabilizer).

### Compactness of the Stabilizer
This is the critical consequence of the action being **proper**.
* In a proper action, the stabilizer $H$ must be a **compact** Lie subgroup of $G$.
* If $H$ were not compact, the action would generally fail to be proper.

### Principal Bundle
In your previous "free" scenario, the bundle was $M \to M/G$.
In this "transitive" scenario, the hierarchy shifts upwards. The group $G$ sits above the manifold $M$, creating a **Principal $H$-Bundle**:
\begin{align}
\pi: G &\to M \\
g &\mapsto g \cdot z_0
\end{align}
* **Total Space:** The Lie group $G$.
* **Base Space:** The manifold $M$ (identified with $G/H$).
* **Fiber:** The compact subgroup $H$.

### Summary Comparison

| Feature | Free Action (Previous) | Transitive Action (Current) |
| :--- | :--- | :--- |
| **Orbit** | A submanifold inside $M$ ($\cong G$) | The entire manifold $M$ |
| **Stabilizer ($H$)** | Trivial ($\{e\}$) | A compact Lie subgroup |
| **Quotient ($M/G$)** | A smooth manifold | A single point (singleton) |
| **Bundle Structure** | $M \to M/G$ is a $G$-bundle | $G \to M$ is an $H$-bundle |

### Why this matters
This structure is the foundation of **Gauge Theory** and **Particle Physics**.
* $M$ represents the spacetime (or a specific geometry like a sphere $S^2 = SO(3)/SO(2)$).
* $G$ represents the total symmetry of the system.
* $H$ represents the "internal" symmetry remaining at a point.

## Masking using Group actions

\begin{align}
\Phi : G \times \mathbb{R}^m \to \mathbb{R}^m,
\end{align}
so that for each $g \in G$, $\Phi$ acts smoothly on the Euclidean state space.
Then the masking can be defined as
\begin{align}
M_{ij} &= \Phi_{g_{ij}^{-1}} (z_j),
\end{align}
and decoding as
\begin{align}
z_j &= \Phi_{g_{ij}} (M_{ij}).
\end{align}

---

##  Mathematical properties

* It **doesn't require any embedding** instead,  $\mathbb{R}^m$ itself carries the structure of a smooth $G$-space.
* The only requirement is that the group action be **free and transitive** (or at least locally invertible in $g$), so that applying the inverse group element undoes the masking.
* All operations remain in $\mathbb{R}^m$, avoiding the topological inconsistencies of trying to globally inject $\mathbb{R}^m$ into a compact group.

* Each combination of $G$ and its action on $\mathbb{R}^m$ defines a **different encryption structure** — with its own balance of interpretability, reversibility, and computational cost.

## **Ex#1-Linear and affine actions**

### (a)  Orthogonal rotations — $G = SO(m)$

Action:
\begin{align}
\Phi_R(z) = R z.
\end{align}

**Properties**

* Deterministic, globally invertible, preserves Euclidean norm.
* Each communication link corresponds to a random coordinate frame.
* Eavesdropper cannot compare across edges (each link sees (R_{ij}z_j)).
* Invariant quantities (e.g., (|z_j|)) are leaked — so useful for “shape-invariant” privacy.

**Use case:**
Obfuscation while preserving energy or distance — e.g., in robotic formations, where direction should be private but scale public.

### General linear group — $G = GL(m,\mathbb{R})$

Action:
\begin{align}
\Phi_A(z) &= A z.
\end{align}

**Properties**

* Adds scaling, shearing, etc.
* Still globally invertible.
* Makes all linear structure indistinguishable to eavesdropper.

**Use case:**
Privacy-preserving *linear consensus* where coordinate transformations destroy any meaning of components or ratios.

---

### Affine transformations — $G = GL(m,\mathbb{R}) \ltimes \mathbb{R}^m $

Action:
\begin{align}
\Phi_{(A,b)} (z) &= A z + b.
\end{align}

**Properties**

* Combines rotation/shear and translation.
* Nonabelian group with a natural semidirect structure.
* Decoding via $A^{-1}(z - b)$.
* Introduces both multiplicative and additive masking.

**Use case:**
A “geometric one-time pad” — you can hide both orientation and origin of each agent's state space.

### Scaling group — $G = (\mathbb{R}^+, \cdot)$

Action:
\begin{align}
\Phi_a(z) = a z.
\end{align}

**Properties**

* Multiplicative obfuscation: hides magnitude relationships.
* Simple, commutative, 1-dimensional group.
* Decoding is trivial $(z = a^{-1} M)$.

**Use case:**
When you want to hide all state magnitudes (e.g., distributed averaging of positive quantities) but maintain ratios.

$G = \mathrm{Conf}(\mathbb{R}^m)$: generated by translations, dilations, rotations, and inversions.

Action:
\begin{align}
\Phi_g(z) = \frac{A z + b}{c^\top z + d},\\ \qquad
\begin{bmatrix} A & b \\ c^\top & d \end{bmatrix} \in O(m+1,1).
\end{align}

**Properties**

* Very rich nonlinear transformations.
* Highly expressive obfuscation — even preserves angles but not distances.
* Smooth, invertible on suitable domains.

**Use case:**
Applications involving *geometry-aware privacy*, e.g. hiding position information while maintaining directional relationships.

---

### Symplectic group — $G = Sp(2n, \mathbb{R})$

Action on phase space $\mathbb{R}^{2n}$:
\begin{align}
\Phi_S((q,p)) = S (q,p).
\end{align}

**Properties**

* Preserves symplectic form (area/volume in phase space).
* Natural for mechanical or Hamiltonian systems.
* Hides individual states but preserves invariants (energy surfaces).

**Use case:**
Masking trajectories of physical systems (robotics, oscillators) while keeping dynamics structure intact.

### Euclidean group — $G = SE(m) = SO(m)\ltimes \mathbb{R}^m$

Action:
\begin{align}
\Phi_{(R,p)}(z) = R z + p.
\end{align}

**Properties**

* Global diffeomorphism on $\mathbb{R}^m$.
* Natural mechanical interpretation — rotation + translation.
* Compact rotational part gives bounded “key space.”

**Use case:**
Privacy-preserving *rigid-body coordination*: each agent's transmitted position appears as if viewed from a random rigid-body transform.

## **Projective and homogeneous-space actions**

### Projective linear group — $G = PGL(m+1,\mathbb{R})$

Action on projective space $\mathbb{RP}^{m}$:
\begin{align}
\Phi_{([A], [z])} = [A z].
\end{align}

**Properties**

* Encodes states up to a scale factor.
* Nonlinear in coordinates, but natural in homogeneous models.
* Provides obfuscation where only *ratios* matter.

**Use case:**
Consensus over proportions (e.g., distributed normalization, probabilities, or concentration ratios).

### Möbius transformations — $G = SL(2,\mathbb{C})$ acting on $\mathbb{C}$

Action:

= \frac{a z + b}{c z + d}.

**Properties**

* Nonlinear, conformal, invertible wherever (c z + d \neq 0).
* Group action on the Riemann sphere.

**Use case:**
For scalar consensus or opinion dynamics, provides a “complex-plane privacy transform” with strong mixing properties.

# Multi-Agent Systems on Lie Groups with Symmetry

## Equivariant systems

Consider a Lie group $G$ with Lie algebra $\mathfrak{g}$ and a subgroup $H\subset G$ with Lie algebra $\mathfrak{h}\subset \mathfrak{g}$.


Assume that the gorup $H\subset G$ acts on $G$ in a smooth and proper manner:
\begin{align}
\Phi : H\times G \to G.
\end{align}
$\Phi$ can be a left, right or a conjugation action.
Let
\begin{align}
\pi : G\to G/H
\end{align}
be the projection map that takes $g\in G$ to its $H$ orbit through $g$. That is $\pi(g)=[g]$.
Note that
\begin{align}
\pi^{-1}([g]) \simeq H/H_g,\qquad H_g=\{h\in H | \Phi_{h}(g)=g\}.
\end{align}

Consider the discrete time system on the Lie group $G$:
\begin{align}
g_{k+1}&=f(g_{k},\xi_{k}), \qquad \xi_k\in \mathfrak{g}.
\end{align}
Note that $\mathfrak{g}$ will either be the space of left-invariant vector fields on $G$ or the left-invariant vector fields on $G$. Recall that if $\xi\in \mathfrak{g}$ is a right vector then the corrsponding left left vector is uniquely given by $\Xi=\mathrm{Ad}_{g^{-1}}\xi$.

We assume that the system dynamics are $H$-equivariant. That is
\begin{align}
f\left(\Phi_h(g),\xi\right)=\Phi_h\left(f\left(g,\xi\right)\right)\qquad \forall \: h\in H.
\end{align}
This implies that the reduced map $\widetilde{f}: G/H \to G/H$ defined by
\begin{align}
\widetilde{f}([g],\xi)=\pi\left(f(g,\xi)\right).
\end{align}
is well defined.
Then the reduced dynamics defined by is well defined.

## Local sections and co-cycles

Define a local section
\begin{align}
\tau : G/H \to G,\qquad \pi \circ \tau = \mathrm{id}_{G/H}
\end{align}
Let $h_{k}\in H$ be such that
\begin{align}
g_{k}&=\Phi_{h_{k}}(\tau\left([g_{k}]\right)).
\end{align}

Note that
\begin{align}
g_{k+1}&=\Phi_{h_{k+1}}\left(\tau\left([g_{k+1}]\right)\right),\\
&=f(g_k,\xi_k)=f\left(\Phi_{h_{k}}(\tau\left([g_{k}]\right)),\xi_k\right)\\
&=\Phi_{h_{k}}\left(f\left(\tau\left([g_{k}]\right),\xi_k\right)\right)
\end{align}
Thus
\begin{align}
\Phi_{h_{k+1}}\left(\tau\left([g_{k+1}]\right)\right)
&=\Phi_{h_{k}}\left(f\left(\tau\left([g_{k}]\right),\xi_k\right)\right).
\end{align}
Thus there exists a $c([g_{k}],\xi_k)\in H$ such that
\begin{align}
f\left(\tau\left([g_{k}]\right),\xi_k\right)&=\Phi_{c([g_{k}],\xi_k)}\left(\tau\left([g_{k+1}]\right)\right).
\end{align}
This defines a map
\begin{align}
c: G/H \times \mathfrak{g} \to H
\end{align}
that is called a co-cycle associated with the section $\tau$.

Note that the $H$-equivariance implies that $[g_{k+1}]=[g_{k}]$ if and only if $\xi_k\in \mathfrak{h}$. That is the dynamics are along the fibre if and only if $\xi_k\in \mathfrak{h}$.

From
\begin{align}
\Phi_{h_{k+1}}\left(\tau\left([g_{k+1}]\right)\right)
&=\Phi_{h_{k}}\circ\Phi_{c([g_{k}],\xi_k)}\left(\tau\left([g_{k+1}]\right)\right),
\end{align}
we see that for a right action
\begin{align}
h_{k+1}
&\in H_{\tau\left([g_{k+1}]\right)}\cdot c([g_{k}],\xi_k)h_{k}
\end{align}
and for a left action
\begin{align}
h_{k+1}
&\in h_{k} c([g_{k}],\xi_k)\cdot H_{\tau\left([g_{k+1}]\right)}
\end{align}

Therefor the the evolution of
\begin{align}
g_{k+1}&=f(g_k,\xi_k)
\end{align}
is uniquely give by
\begin{align}
[g_{k+1}]&=\widetilde{f}\left([g_{k}],\xi_k\right),\\
g_{k+1}&=\Phi_{h_{k+1}}\left(\tau([g_{k}])\right)
\end{align}
Note that the re-cosntruction from the reduced base dynamics entirely depend on the choice of the section $\tau$.

For the special case of free left multiplication actions  we have
\begin{align}
c([g_{k}],\xi_k)&=f\left(\tau\left([g_{k}]\right),\xi_k\right)\left(\tau\left([g_{k+1}]\right)\right)^{-1}\\
h_{k+1}
&=h_{k}c([g_{k}],\xi_k),
\end{align}
and for the special case of free right multiplication actions we have
\begin{align}
c([g_{k}],\xi_k)&=\left(\tau\left([g_{k+1}]\right)\right)^{-1}f\left(\tau\left([g_{k}]\right),\xi_k\right)\\
h_{k+1}
&=c([g_{k}],\xi_k)h_{k},
\end{align}

Consider the case where in addition the dynamics are simply kinematic systsms. That is when $f(g_k,\xi_k)=\exp{(\xi_k)}g_k=g_k\exp{(\Xi_k)}$ where $\xi_k=\mathrm{Ad}_{g_k}(\Xi_k)$ where $\xi_k\in \mathfrak{g}$ is a right velocity. Then
when $\Phi$ is simply left multiplication, using the left invariant kinematics, we find that
\begin{align}
c([g_{k}],\Xi_k)&=\tau\left([g_{k}]\right)\,\exp{\left(\Xi_k\right)}\,\tau\left([g_{k+1}]\right)^{-1}, \qquad \Xi_k\in\mathfrak{g}
\end{align}
and if $\Phi$ is right multiplication, using the right-invariant kinematics, we find that
\begin{align}
c([g_{k}],\xi_k)&=\tau\left([g_{k+1}]\right)^{-1}\,\exp{\left(\xi_k\right)}\,\tau\left([g_{k}]\right), \qquad \xi_k\in\mathfrak{g}.
\end{align}
Note that when $\Phi$ is a left action we use left velocites and when $\Phi$ is a right action we use right velocites to represent the kinematics.

Now if we restrict $\xi_k^h\in \mathfrak{h}$ then in the special case where the action $\Phi$ is simply left multiplication we find that
\begin{align}
c([g_{k}],\Xi_k^h)&=\exp{\left(\mathrm{Ad}_{\tau\left([g_{k}]\right)}\Xi_k^h\right)}, \qquad \Xi_k^h\in\mathfrak{h}
\end{align}
and if the action is right multiplication we find that
\begin{align}

\end{align}

T_g G &\simeq \mathrm{Ker}(d\pi)\oplus \mathscr{H}_g
\end{align}



A section, $\tau : G/H \to G$ such the $\pi \circ \tau=\mathrm{id}_{G/H}$ where $d\tau_{[g]} (T_{[g]}(G/H))\subseteq \mathscr{H}_{\tau([g])}$ is horizontal if and only if
\begin{align}
T_{\tau([g])}L_{\tau([g])^{-1}}, d\tau_{[g]}(v) \in \mathfrak{m},
\qquad \forall, v \in T_{[g]}(G/H).
\end{align}
Equivalently,
\begin{align}
\mathrm{Proj}_{\mathfrak{h}}\left(\tau^*\theta_L\right)=0
\end{align}
where
$\theta_L=g^{-1}dg$ is the left Maurer-Cartan form on $G$.
Such a section always exists locally.
For such a section one can also show that for left actions
\begin{align}
c([g_{k}],\Xi_k^m)=e,\qquad \forall \quad \Xi_k^m\in\mathfrak{m},
\end{align}
and for right actions
\begin{align}
c([g_{k}],\xi_k^m)=e,\qquad \forall \quad \xi_k^m\in\mathfrak{m}.
\end{align}




On a reductive $(G,H)$ space, for right H-actions and $H$-equivariant sections $\tau : G/H \to G$ the reduced and re-cosntruction dynamics of
\begin{align}
g_{k+1}&=\exp{(\xi_k^h+\xi_k^m)}\,g_k\qquad \xi_k^m\in \mathfrak{m}, \xi_k^h\in \mathfrak{h}
\end{align}
are given by
\begin{align}
[g_{k+1}]&=\widetilde{f}([g_{k}],\xi_k^m),\qquad \xi_k^m\in \mathfrak{m},\\
h_{k+1}&=\exp{\left(\mathrm{Ad}_{\tau\left([g_{k}]\right)^{-1}}\xi_k^h\right)}\,h_{k},\qquad \xi_k^h\in \mathfrak{h},\\
g_{k+1} &=\Phi_{h_{k+1}}(\tau[g_{k+1}]).
\end{align}




## Example of the Hopf Fibration - $SO(2)$ version

\begin{align}
G &= \mathrm{SO}(3), \quad H = \mathrm{SO}(2),
\end{align}
and $H$ acts on $G$ by rotation about a fixed axis (say, the **z-axis**).
This is a canonical case of symmetry reduction on a Lie group.

This action is also free.

---

### 1. The setting

* The Lie group $G = \mathrm{SO}(3)$ is the set of all rotation matrices in 3D.
* Its Lie algebra is $\mathfrak{g} = \mathfrak{so}(3)$, the space of **skew-symmetric** $3\times3$ matrices, which can be identified with $\mathbb{R}^3$ via the hat map $\hat{\cdot}$.
* The subgroup $H = \mathrm{SO}(2)$ consists of all rotations **around the z-axis**, i.e.
\begin{align}
H = \left\{ R_z(\theta) =
\begin{bmatrix}
\cos\theta & -\sin\theta & 0 \\
\sin\theta & \cos\theta & 0 \\
0 & 0 & 1
\end{bmatrix}
\right\}.
\end{align}

The action of $H$ on $G$ is by **right multiplication**:
\begin{align}
\Phi_h(g) = gh.
\end{align}

---

### 2. The quotient space $G/H$

Since $H$ stabilizes the **z-axis**, the quotient
\begin{align}
G/H \cong S^2,
\end{align}
the **2-sphere**, represents the *direction of the body’s z-axis in space*.
The projection map
\begin{align}
\pi : \mathrm{SO}(3) \to S^2, \quad \pi(R) = R e_3,
\end{align}
sends a rotation matrix $R$ to the image of the unit direction $e_3 = [0,0,1]^T$.
Thus, the reduced variable is
\begin{align}
[R] = R e_3 \in S^2.
\end{align}

---

### 3. Agent dynamics on $\mathrm{SO}(3)$

Let each agent $i$ evolve as
\begin{align}
R_{i,k+1} = \exp\left(\widehat{\omega}_{i,k}\right)R_{i,k},
\end{align}
where $\omega_{i,k} \in \mathbb{R}^3$ is the spatial angular velocity (control input),
and $\widehat{\omega}_{i,k}$ is the skew-symmetric matrix associated with $\omega_{i,k}$.


---

### 4. $H$-equivariance

Right multiplication by any $h \in SO(2)$ commutes with the dynamics:
\begin{align}
f(Rh, u) = \exp(\widehat{u})Rh = (\exp(\widehat{u})R)h = \Phi_h(f(R,u)).
\end{align}
Thus $f$ is $H$-equivariant.

---

### 5. Reduction to $G/H \cong S^2$

The reduced variable is $r_{i,k} = R_{i,k} e_3 \in S^2$.

Then
\begin{align}
r_{i,k+1} = R_{i,k+1} e_3 = \exp(\widehat{\omega}_{i,k}) R_{i,k}e_3
= \exp(\widehat{\omega}_{i,k})r_{i,k}
=\widetilde{f}(r_{i,k}, \Omega_{i,k}),
\end{align}
which defines a discrete-time system on the sphere $S^2$.

The map $\widetilde{f}$ is precisely
\begin{align}
\widetilde{f}(r, \omega) = \pi(\exp(\widehat{\omega})R)=\exp(\widehat{\omega})r.
\end{align}

---

### 6. Local section (reconstruction)

A **local section** $\tau: S^2 \to SO(3)$ can be chosen as follows:
for a given $r \in S^2$, let $\tau(r)$ be the unique rotation matrix that maps $e_3 \mapsto r$

Let
\begin{align}
b_1 = \frac{e_3 \times r}{|e_3 \times r|}, \qquad b_2 = b_1 \times r.
\end{align}
Then define
\begin{align}
\tau(r) = \begin{bmatrix} b_1 & b_2 & r\end{bmatrix},
\end{align}

Then given the reduced trajectory $r_k \in S^2$, the full rotation is reconstructed as
\begin{align}
R_{k+1} = \Phi_{h_{i,k+1}}\left(\tau(r_{k+1})\right),
\end{align}
where $h_{i,k+1}\in H$ evolves as:
\begin{align}
h_{i,k+1}&=\tau(r_{k})^{-1}\exp{(\xi_k)}\;\tau(r_{k})h_{i,k}=\exp{\left(\mathrm{Ad}_{\tau(r_{k})^{-1}}\xi_k\right)}\;h_{i,k}
\end{align}

---

### 7. Consensus interpretation

* On $S^2$, the consensus means
\begin{align}
r_{i,k} \to r^*, \quad \forall i.
\end{align}
* This is *attitude alignment modulo rotation about the z-axis*.
* Without additional synchronization in $H$, each $R_i$ can differ by a residual rotation about $r^*$.


If each agent implements a fibre update by appropriately choosing
$\xi_{i,k}\in\mathfrak{so}(2)$
such that $\lim_{k\to\infty}h_{i,k}=h^*_i$
then,
\begin{align}
R_{i,k} \to \tau (r^*)h_i^* , \quad h_i^* \in SO(2),
\end{align}
and if the agents also synchronize their yaw angles (the $SO(2)$-part), then
\begin{align}
R_{i,k} \to \tau(r^*)h^* , \quad \forall i.
\end{align}

## Hopf Fibration - $SU(2)$ double cover

### Hopf fibration setting

\begin{align}
G &= SU(2), \quad H = U(1),
\end{align}
and (H) acts on (G) **by right multiplication**:
\begin{align}
\Phi_h(q) = qh, \quad h \in U(1).
\end{align}
This is the **Hopf fibration**
\begin{align}
U(1) \hookrightarrow SU(2) \xrightarrow{\pi} S^2.
\end{align}

---

### 1. The setting

* The Lie group $G = SU(2)$ is the group of $2\times 2$ complex unitary matrices with determinant 1:
\begin{align}
SU(2) = \left\{ q \in \mathbb{C}^{2\times2} \,\middle|\, q^\dagger q = I,\ \det(q)=1 \right\}.
\end{align}
* Its Lie algebra is
\begin{align}
\mathfrak{g} = \mathfrak{su}(2) = { X \in \mathbb{C}^{2\times 2} \mid X^\dagger = -X, \ \mathrm{tr}(X)=0 }.
\end{align}
  A convenient basis is the Pauli matrices:
\begin{align}
\sigma_1 =
\frac{1}{2}\begin{bmatrix}0 & 1\\ 1 & 0\end{bmatrix}, \quad
\sigma_2 =
\frac{1}{2}\begin{bmatrix}0 & -i\\ i & 0\end{bmatrix}, \quad
\sigma_3 =
\frac{1}{2}\begin{bmatrix}1 & 0\\ 0 & -1\end{bmatrix}.
\end{align}

* The subgroup $H = U(1)$ consists of complex phase rotations:
\begin{align}
H = \left\{ e^{i\theta\sigma_3} = \mathrm{diag}(e^{i\frac{\theta}{2}}, e^{-i\frac{\theta}{2}}) \mid \theta \in \mathbb{R} \right\}.
\end{align}
This corresponds to rotation about the **z-axis**.

The action of $H$ on $G$ is
\begin{align}
\Phi_h(q) = qh, \qquad h \in U(1).
\end{align}

---

### 2. The quotient space $G/H \simeq S^2$

The quotient
\begin{align}
SU(2)/U(1) \cong S^2
\end{align}
is the **base manifold** of the Hopf fibration.
The projection map
\begin{align}
\pi: SU(2) \to S^2, \qquad \pi(q) = q \sigma_3 q^\dagger,
\end{align}
maps each $q \in SU(2)$ to a point on the 2-sphere (the **Bloch vector**).

In coordinates, this gives a point
\begin{align}
r = \pi(q) \in S^2 \subset \mathbb{R}^3, \quad
r = (r_1, r_2, r_3), \quad r_1^2 + r_2^2 + r_3^2 = 1,
\end{align}
where
\begin{align}
r_k = \frac{1}{2}\mathrm{Tr}(q \sigma_3 q^\dagger \sigma_k), \quad k=1,2,3.
\end{align}
Thus $r$ represents the **direction of the “body z-axis”**, just as before.

---

### 3. Agent dynamics on $SU(2)$

Let each agent evolve as
\begin{align}
q_{i,k+1} = \exp\left(-i \omega_{i,k}\cdot \sigma \right) q_{i,k},
\end{align}
where $\omega_{i,k} \in \mathbb{R}^3$ is the spatial angular velocity,
and $\sigma = (\sigma_1, \sigma_2, \sigma_3)$.

Explicitly the exponential map is given by
\begin{align}
\exp\left(-i\omega\cdot\sigma\right)
&= \cos\left(\tfrac{||\omega||}{2}\right) I - 2i\sin\left(\tfrac{||\omega||}{2}\right)
\frac{\omega}{||\omega||}\cdot\sigma.
\end{align}


This is the **discrete-time kinematic system** on $SU(2)$.
The corresponding map on $SO(3)$ is $R_{i,k+1} = \exp(\widehat{\omega}_{i,k}) R_{i,k}$.

---

### 4. H-equivariance

Right multiplication by any $h \in U(1)$ commutes with the dynamics:
\begin{align}
f(qh, u) = \exp\left(-i u\cdot\sigma\right) qh
= \left(\exp\left(-iu\cdot\sigma\right) q\right) h
= \Phi_h(f(q,u)).
\end{align}
Thus the system is $U(1)$-equivariant.

---

### 5. Reduction to $G/H \cong S^2$

The reduced variable is
\begin{align}
r_{i,k} = \pi(q_{i,k}) = q_{i,k} \sigma_3 q_{i,k}^\dagger \in S^2.
\end{align}

Then

\begin{align}
r_{i,k+1}
&= q_{i,k+1}\sigma_3 q_{i,k+1}^\dagger
= \exp\left(-i\omega_{i,k}\cdot\sigma\right) q_{i,k}\sigma_3 q_{i,k}^\dagger \exp\left(i\omega_{i,k}\cdot\sigma\right) \\
&= \exp(-i\omega_{i,k}\cdot\sigma) r_{i,k} \exp(i \omega_{i,k}\cdot\sigma).
\end{align}
Using the adjoint action $\mathrm{Ad}_{\exp(-i \omega\cdot\sigma)}$, this is equivalent to the rotation on $S^2$:
\begin{align}
r_{i,k+1} = \exp(\widehat{\omega}_{i,k}) r_{i,k}.
\end{align}

Thus the induced reduced map is
\begin{align}
\widetilde{f}(r, \omega) = \exp(\widehat{\omega})r.
\end{align}

---

### 6. Local section (reconstruction)

A **local section** $\tau: S^2 \to SU(2)$ assigns to each $r \in S^2$ a unit quaternion (or $SU(2)$ matrix) that rotates the north pole $\sigma_3$ to $r$.

One choice (Hopf section) is
\begin{align}
\tau(r) =
\begin{bmatrix}
\cos(\tfrac{\theta}{2}) e^{-i\phi/2} & -\sin(\tfrac{\theta}{2}) e^{-i\phi/2} \\
\sin(\tfrac{\theta}{2}) e^{i\phi/2} & \cos(\tfrac{\theta}{2}) e^{i\phi/2}
\end{bmatrix},
\end{align}
where $r = (\sin\theta\cos\phi, \sin\theta\sin\phi, \cos\theta)$.

Given the reduced trajectory $r_{i,k} \in S^2$, the full state is reconstructed as
\begin{align}
q_{i,k+1} = \tau(r_{i,k+1}) h_{i,k+1},
\end{align}
where
\begin{align}
h_{i,k+1} &= \tau(r_{i,k})^{-1}\exp\left(-i \psi_{i,k}\sigma_3 \right) \tau(r_{i,k}) h_{i,k}=
\exp\left(-i \psi_{i,k}\mathrm{Ad}_{\tau(r_{i,k})^{-1}}\sigma_3 \right)  h_{i,k}, \qquad \psi_{i,k} \in \mathbb{R},\\
&=
\exp\left(i \psi_{i,k}r_{i,k}\cdot\sigma \right)  h_{i,k}, \qquad \psi_{i,k} \in \mathbb{R},\\
\end{align}

Thus the full evolution is
\begin{align}
r_{i,k+1} &= \exp(\widehat{\omega}_{i,k}) r_{i,k},\\
q_{i,k+1} &= \tau(r_{i,k+1})\exp\left(i \psi_{i,k}r_{i,k}\cdot\sigma \right)  h_{i,k},
\end{align}

---

### 7. Consensus interpretation

* On $S^2$:
consensus means all reduced directions align,
\begin{align}
r_{i,k} \to r^*, \quad \forall i.
\end{align}

* On $H$:
Each agent chooses $\psi_{i,k}\in \mathbb{R}$
such that
\begin{align}
h_{i,k} \to h^*_i,
\end{align}

* On $SU(2)$:
each $q_i$ may differ by a residual **phase** $h_i^* \in U(1)$:
\begin{align}
q_{i,k} \to \sigma(r^*) h_i^*, \quad h_i^* \in U(1).
\end{align}

This represents *directional consensus modulo a phase rotation* in the $U(1)$ fibre — the **Hopf-fibre analogue** of the yaw ambiguity in $SO(3)$.

If the agents also synchronize their fibre phases, $h_i^*$, then
\begin{align}
q_{i,k} \to \sigma(r^*)h^*, \quad \forall i,
\end{align}
yielding full $SU(2)$ synchronization.


## Fibre Convergence

Let each agent's fibre element be parameterized by a phase $\psi_{i,k}\in\mathbb{S}$ as
\begin{align}
\bar{h}_{i,k} &= \tau(r_{i,k})e^{-i\psi_{i,k}\sigma_3}\tau(r_{i,k})^{-1},
\end{align}
and let the desired target be
\begin{align}
\bar{h}^* &= \tau(r^*)e^{-i\psi^*\sigma_3}\tau(r^*)^{-1}.
\end{align}

We derive a simple convergence law for $\bar{h}_{i,k}$ from a **Morse-type potential** on the circle.

---

### A Morse potential on $U(1)$

Define a smooth potential function with a unique minimum at $\bar{h}=\bar{h}^*$:
\begin{align}
V(\bar{h}) &= 1 - \tfrac{1}{2}\mathrm{Tr}\left((\bar{h}^*)^\dagger \bar{h}\right).
\end{align}

Compute explicitly:
\begin{align}
\mathrm{Tr}\left((\bar{h}^*)^\dagger \bar{h}\right)
&\approx\mathrm{Tr}\left(e^{i\psi^*\sigma_3}e^{-i\psi\sigma_3}\right)
= \mathrm{Tr}\left(e^{-i(\psi-\psi^*)\sigma_3}\right)
= 2\cos(\psi-\psi^*),
\end{align}
so
\begin{align}
V(\psi) &= 1 - \cos(\psi - \psi^*).
\end{align}

This is the classic **polar Morse potential** on the circle:

* $V \ge 0$,
* $V = 0$ iff $\psi = \psi^*$,
* smooth everywhere except for the antipodal instability at $\psi = \psi^* + \pi$.

---

### Gradient descent on (U(1))

The gradient is
\begin{align}
\frac{\partial V}{\partial \psi} = \sin(\psi - \psi^*),
\end{align}
and a discrete-time descent iteration is
\begin{align}
\psi_{k+1} = \psi_k - \alpha\sin(\psi_k - \psi^*), \qquad \alpha > 0.
\end{align}

---

### Group update (lifted to the fibre)

In group form, the update on (U(1)) is
\begin{align}
\bar{h}_{k+1} = \exp\big(-i\alpha\sin(\psi_k - \psi^*)\sigma_3\big)\bar{h}_k.
\end{align}

To express this in the full group (SU(2)), we use the local section (\tau(r_k)) to lift the fibre motion:
\begin{align}
h_k &= \tau(r_k)^{-1}\bar{h}_k\tau(r_k).
\end{align}

Then the corresponding **conjugated fibre update** in the total space is
\begin{align}
h_{k+1}
&= \tau(r_k)^{-1}\exp\big(-i\alpha\sin(\psi_k - \psi^*)\sigma_3\big)\tau(r_k)h_k,\\
&=\exp\big(i\alpha\sin(\psi_k - \psi^*)r_k\cdot\sigma\big)h_k,
\end{align}

This is the right-invariant lifted gradient flow along the (U(1)) fibre over the base point (r_k).

---

### Final discrete polar Morse update

\begin{align}
\psi_{i,k+1} &= \psi_{i,k} - \alpha\sin(\psi_{i,k} - \psi^*),\\
h_{i,k+1} &= \tau(r_{i,k})^{-1}
\exp\big(-i\alpha\sin(\psi_{i,k} - \psi^*)\sigma_3\big)
\tau(r_{i,k}),h_{i,k}.
\end{align}

---

### Geometric interpretation

* $V = 1 - \cos(\psi - \psi^*)$ is the **Morse potential** on the fibre $U(1)$.
* The update follows its **negative gradient**.
* The conjugation by $\tau(r_{i,k})$ aligns the vertical motion with the fibre direction over the current base point $r_{i,k}$.
* The generator $-i\sin(\psi_{i,k}-\psi^*),\tau(r_{i,k})\sigma_3\tau(r_{i,k})^{-1}$ defines a **vertical velocity** in $\mathfrak{su}(2)$.
* Convergence is global except for the antipodal singularity at $\psi=\psi^*+\pi$.

---

## Discrete-Time Projective Reduction of $G=SE(3)$ with $H=SO(3)$: Image-Plane Dynamics and Reconstruction

\begin{align}
SE(3) \xrightarrow{H = SO(3)} \mathbb{P}^2.
\end{align}

---

### 1. **Full dynamics on (SE(3))**

Let each camera pose be
\begin{align}
g_k =
\begin{bmatrix}
R_k & t_k \\
0 & 1
\end{bmatrix},
\qquad R_k \in SO(3), \qquad t_k \in \mathbb{R}^3.
\end{align}

Discrete-time rigid motion update:
\begin{align}
g_{k+1} = g_k \exp(\xi_k),
\end{align}
with
\begin{align}
\xi_k =
\begin{bmatrix}
\widehat{\Omega}_k & v_k \\
0 & 0
\end{bmatrix} \in \mathfrak{se}(3).
\end{align}

Expanding:
\begin{align}
R_{k+1} = R_k \exp(\widehat{\Omega}_k), \qquad
t_{k+1} = t_k + R_k V_k,
\end{align}
where $V_k$ is the body-frame translation over the step (approximating $v_k \Delta t$).

---

### 2. **Symmetry action**

Let $H = SO(3)$ act on $G = SE(3)$ by **left multiplication**:
\begin{align}
\Phi_h(g) = h g = (hR, ht).
\end{align}

This corresponds to changing the inertial frame by a rotation $h$.
Since image formation depends only on *relative geometry*, this symmetry preserves the projection.

---

### 3. **Quotient projection**

Define the projection map
\begin{align}
\pi : SE(3) \to \mathbb{P}^2, \quad
\pi(R,t) = [R^\top (X - t)],
\end{align}
for a fixed world point (X).
The homogeneous image coordinates $x_k = \pi(g_k)$ live on the projective plane $\mathbb{P}^2$.

This map satisfies $\pi(hg) = \pi(g)$ for all $h \in SO(3)$.

---

### 4. **Reduced dynamics on the image plane**

We now compute how $x_k$ evolves under the discrete update $g_{k+1} = g_k \exp(\xi_k)$.

Let $x_k = R_k^\top (X - t_k)$ (normalized to $\mathbb{P}^2$).
Then
\begin{align}
x_{k+1} = R_{k+1}^\top (X - t_{k+1})
= \exp(-\widehat{\Omega}_k) R_k^\top \bigl(X - t_k - R_k V_k\bigr).
\end{align}
Substitute $R_k^\top (X - t_k) = x_k$:
\begin{align}
x_{k+1} = \exp(-\widehat{\Omega}_k) (x_k - V_k).
\end{align}

This gives the **reduced discrete-time update** on $\mathbb{P}^2$:
\begin{align}
x_{k+1} = \exp(-\widehat{\Omega}_k) (x_k - V_k)
\end{align}
which is the discrete analogue of the continuous optical-flow equation
\begin{align}
\dot{x} = -\widehat{\Omega}x - R^\top v.
\end{align}

---

### 5. **Interpretation**

* $x_k$ is the *direction* of the observed point in the camera frame (image-plane coordinate).
* The term $\exp(-\widehat{\Omega}_k)$ accounts for **rotational motion** of the camera.
* The subtraction $(x_k - V_k)$ corresponds to **translational parallax**.

This update rule lives entirely in the **quotient space** $G/H$ the camera's image space.

---

### 6. **Reconstruction**

To reconstruct the full pose $g_k = (R_k, t_k)$ from the reduced trajectory $x_k$:

1. Integrate the rotational motion:
\begin{align}
R_{k+1} = R_k \exp(\widehat{\Omega}_k).
\end{align}
2. Integrate translation:
\begin{align}
   t_{k+1} = t_k + R_k V_k.
\end{align}
3. If we define a canonical local section
\begin{align}
\sigma(x_k) = \text{“rotation whose third column aligns with } x_k”,
\end{align}
then we can write
\begin{align}
g_k = h_k , \sigma(x_k),
\end{align}
where $h_k \in SO(3)$ is the **fiber variable** describing roll around the viewing axis.

The fiber evolves by the **reconstruction equation**:
\begin{align}
h_{k+1} = h_k \exp(\widehat{\Omega}_{k,\parallel}),
\end{align}
where $\omega_{k,\parallel} = (\omega_k \cdot x_k) x_k$
is the component of angular velocity along the viewing direction —
the same as the “phase evolution” in the discrete Hopf fibration.

---


### In summary

* The **reduced dynamics** on $G/H$ describe **how features move on the camera screen** in discrete time.
* The **fiber** $SO(3)$ encodes the camera's **absolute orientation**, which can be reconstructed if the rotational component of motion is known.
* The resulting discrete-time relationship:
\begin{align}
x_{k+1} = \exp(-\widehat{\Omega}_k)(x_k - V_k)
\end{align}
is the projective-geometric analogue of the Hopf fibration reduction and reconstruction pattern.

---
